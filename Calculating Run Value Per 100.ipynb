```python
#importing libraries
import pandas as pd
import pickle
from pandas import DataFrame, Series
import numpy as np
import seaborn as sns
import statsmodels.api as sm
import matplotlib.pyplot as plt
import pybaseball as pyb
from pybaseball import statcast
```

```python
df = pd.read_csv("/Users/johndavis/Desktop/MLBRunValues.csv",low_memory=False)
```

```python
#Drop all rows with NaN values
df = df.replace([np.inf, -np.inf], np.nan)
df = df.dropna()
df = df.reset_index()
```

```python
#Create New dataframes for every seperate pitch type
dff = df[df['Pitch Type'] == 'FF']
dfc = df[df['Pitch Type'] == 'FC']
dsl = df[df['Pitch Type'] == 'SL']
dch = df[df['Pitch Type'] == 'CH']
dcb = df[df['Pitch Type'] == 'CU']
dspl = df[df['Pitch Type'] == 'SPL']
```

```python
#Create a new dataframe for each pitch type that lists pitcher name once the average of velocity, spin rate, vertical break, horizontal break, and release extension and the sum of run values
dff_rv = dff.groupby(['Player Name']).agg({'Velocity': 'mean', 'Spin Rate': 'mean', 'Vertical Break': 'mean','Horizontal Break': 'mean',  'Release Height': 'mean', 'Side Release': 'mean','Extension': 'mean', 'Run Values': 'sum'})
#add a colums that is the total amount of pitches thrown
dff_rv.insert(1, 'Pitch Count', dff.groupby(['Player Name']).size())
# add a column that divides Pitch count by 100
dff_rv.insert(9, 'Pitch Count/100', dff_rv['Pitch Count']/100)
#add a column that divides run values by pitch count/100
dff_rv.insert(9, 'RV/100', dff_rv['Run Values']/dff_rv['Pitch Count/100'])
#delete rows that have less than 100 pitches thrown
dff_rv = dff_rv[dff_rv['Pitch Count'] > 100]
#sort by run value per 100 pitches
dff_rv = dff_rv.sort_values(by=['Run Values'], ascending=True)
dff_rv = dff_rv.reset_index()
dff_rv.head()
```

```python
#Repeat for Slider
dsl_rv = dsl.groupby(['Player Name']).agg({'Velocity': 'mean', 'Spin Rate': 'mean', 'Vertical Break': 'mean','Horizontal Break': 'mean',  'Release Height': 'mean', 'Side Release': 'mean','Extension': 'mean', 'Run Values': 'sum'})
dsl_rv.insert(1, 'Pitch Count', dsl.groupby(['Player Name']).size())
dsl_rv.insert(9, 'Pitch Count/100', dsl_rv['Pitch Count']/100)
dsl_rv.insert(9, 'RV/100', dsl_rv['Run Values']/dsl_rv['Pitch Count/100'])
dsl_rv = dsl_rv[dsl_rv['Pitch Count'] > 100]
dsl_rv = dsl_rv.sort_values(by=['Run Values'], ascending=True)
dsl_rv = dsl_rv.reset_index()
dsl_rv.head()
```

```python
#Repeat for Changeup
dch_rv = dch.groupby(['Player Name']).agg({'Velocity': 'mean', 'Spin Rate': 'mean', 'Vertical Break': 'mean','Horizontal Break': 'mean',  'Release Height': 'mean', 'Side Release': 'mean','Extension': 'mean', 'Run Values': 'sum'})
dch_rv.insert(1, 'Pitch Count', dch.groupby(['Player Name']).size())
dch_rv.insert(9, 'Pitch Count/100', dch_rv['Pitch Count']/100)
dch_rv.insert(9, 'RV/100', dch_rv['Run Values']/dch_rv['Pitch Count/100'])
dch_rv = dch_rv[dch_rv['Pitch Count'] > 100]
dch_rv = dch_rv.sort_values(by=['Run Values'], ascending=True)
dch_rv = dch_rv.reset_index()
dch_rv.head()
```

```python
#Repeat for Curveball
dcb_rv = dcb.groupby(['Player Name']).agg({'Velocity': 'mean', 'Spin Rate': 'mean', 'Vertical Break': 'mean','Horizontal Break': 'mean',  'Release Height': 'mean', 'Side Release': 'mean','Extension': 'mean', 'Run Values': 'sum'})
dcb_rv.insert(1, 'Pitch Count', dcb.groupby(['Player Name']).size())
dcb_rv.insert(9, 'Pitch Count/100', dcb_rv['Pitch Count']/100)
dcb_rv.insert(9, 'RV/100', dcb_rv['Run Values']/dcb_rv['Pitch Count/100'])
dcb_rv = dcb_rv[dcb_rv['Pitch Count'] > 100]
dcb_rv = dcb_rv.sort_values(by=['Run Values'], ascending=True)
dcb_rv = dcb_rv.reset_index()
dcb_rv.head()
```

```python
#Download all dataframes as csv files
dff_rv.to_csv(r'/Users/johndavis/Desktop/MLBRunValuesFF.csv', index = False)
dsl_rv.to_csv(r'/Users/johndavis/Desktop/MLBRunValuesSL.csv', index = False)
dch_rv.to_csv(r'/Users/johndavis/Desktop/MLBRunValuesCH.csv', index = False)
dcb_rv.to_csv(r'/Users/johndavis/Desktop/MLBRunValuesCB.csv', index = False)
```
